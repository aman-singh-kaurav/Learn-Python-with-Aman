**What is Programming language?**

It is nothing but a medium to communicate with computers.

Programming language has :
1) Predefined rules: Syntax<br>
2) Predefined words: Keywords

***

**What is Low Level Language and High Level Language?**

On the basis of ease of Human Understanding:

1) **Low level Language(LLL):** Languages that are difficult for humans to understand.<br>
Ex: Machine language, Assembly language.<br>
2) **High level Language(HLL):** Languages that are easier for humans to understand.<br>
Ex: Python, C, C++, Java.

***

**What is Assembly Language vs Machine Language?**

**Assembly Language:** Intermediate Language between High Level Language and Low Level Language. <br>
**Machine Language:** Low Level Language which uses only 0 (OFF) and 1 (ON).

***

**What is Compiler and Interpreter?**

**Compiler:** Takes the whole code at once and converts it.<br>
If there is error in code, it will not translate anything   


**Interpreter:** Takes the code line by line (Instruction by instruction) and converts it.<br>
If there is error in code, it will translate the cod uptil the line line which has error

***

**Different Types of Errors in Python**

- **Compile Time Errors**
    - Indentation Error
    - Syntax Error (Grammatical Error)
- **Run Time Errors**
    - Logical Errors

***

**Source code->(compiler)->Byte code->(interpreter)->Machine code**

**Source code:** It is the code written by humans.<br>
**Intermediate code:** It is the intermediate code generated by translator(compiler/interpreter).<br>
* Python(Byte code)<br>
* C/C++ (Executable)<br>
* Java (Object code)<br>

**Machine code:** Final code that computer understands. It is binary in nature.
